Objective:
>In this project, you will perform a one-sample Z-test to evaluate a claim about a population mean when the population standard deviation is known.

A coffee company claims that the average caffeine content in one cup of its coffee is 120 mg.

From long-term production records, the company knows that the population standard deviation is 15 mg.

You randomly sample 50 cups of coffee and record their caffeine content as below:


```python
data = [
    118,
    122,
    125,
    117,
    119,
    121,
    123,
    116,
    124,
    120,
    118,
    121,
    126,
    115,
    117,
    123,
    122,
    119,
    124,
    118,
    116,
    120,
    123,
    127,
    121,
    122,
    119,
    124,
    125,
    118,
    117,
    123,
    121,
    120,
    116,
    122,
    123,
    125,
    118,
    121,
    119,
    120,
    117,
    124,
    122,
    118,
    121,
    123,
    120,
    116,
]
```

At a significance level of Î± = 0.05, test whether the true mean caffeine content is different from 120 mg.

Tasks:
* Write a complete Python program that prints:
  1. The sample size (n)
  2. The sample mean
  3. The z-statistic
  4. The p-value

## Task 1: Print Sample Size

```python
sample_size = len(data)
print("Sample size is as follows:", f"n = {sample_size}", sep="\n")
```

## Task 2: Print Sample Mean

```python
sample_mean = sum(data) / len(data)
print("Sample mean is as follows:", f"n = {sample_mean}", sep="\n")
```

## Task 3: Print Z-Statistic

formula =$ z = \frac{\bar{x} - \mu}{\sigma / \sqrt{n}} $



```python
x_bar = sample_mean
mu = 120
sigma = 15
n = sample_size

z_stat = (x_bar - mu) / (sigma / n**0.5)
print("Z-Score is as follows:", f"n = {z_stat}", sep="\n")
```

## Task 4: Print P-Value

In z-table, when Z=0.27, $\phi$(0.27) = 0.60642

Therefore P(Z <= 0.27) = 0.60642

To find right side of this probability, subtract from one. Then multiply by two since it is a two-tailed test.

```python
p_value = 2 * (1 - 0.60642)
print("P-Value is as follows:", f"n = {p_value}", sep="\n")
```

## Task 5: Print Critical Z-Value

$\alpha$ = .05, $\frac{\alpha}{2}$ = 0.025
1 - 0.025 = 0.975

Critical Z-Value for 0.975 = 1.96

P(-1.96 <= z <= 1.96) = 0.95

## Task 6: Caffeine Content Histogram

```python
import matplotlib.pyplot as plt

plt.hist(data)

plt.title("Caffeine Content Histogram")
plt.axvline(x=mu, color="r", label="Claimed Mean")
plt.axvline(x=x_bar, color="g", label="Sample Mean")
plt.legend()
plt.show()
```

## Task 7: Final Decision

### We fail to reject the null hypothesis, since z-statisic is within the range of -1.96 to 1.96 .

Extra: Using scipy

```python
from scipy import stats

# Everything up to calculating Z-Stat is the same
# var z_stat comes from above
alpha = 0.05

# Here we use stats.norm.cdf(abs(z_score))) to find the z-score table value
# Two-sided p-value: 2 * (1 - Î¦(|z|))
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

# Same here for z-critical value with stats.norm.ppf(1 - alpha / 2)
# Critical z-value for two-sided test at alpha = 0.05
# z_(alpha/2) = Î¦^{-1}(1 - alpha/2)
z_crit = stats.norm.ppf(1 - alpha / 2)

# Print results
print(f"Sample size (n): {n}")
print(f"Sample mean: {sample_mean:.4f} mg")
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.6f}")
print(f"Critical z-value (Â±z_crit): Â±{z_crit:.4f}")

# Final decision
if abs(z_stat) > z_crit:
    decision = (
        "Reject H0: there is evidence that the true mean is different from 120 mg."
    )
else:
    decision = "Fail to reject H0: there is not enough evidence that the true mean is different from 120 mg."

print("\nDecision:")
print(decision)
```
