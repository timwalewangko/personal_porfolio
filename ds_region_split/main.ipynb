```python
import numpy as np
import pandas as pd
from modules.forecast import generate_forecast
import matplotlib.pyplot as plt
```

# Step 0: Import Data
### See data here https://tinyurl.com/3j5er69y

```python
# sales data from file
sales = pd.read_csv("data/sales.csv", index_col="week", parse_dates=True)
sales.plot()
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

# Step 1: Generate Forecast

```python
# Use XGBoost to generate time series forecast
forecast_a, importance_plot_a, errors_a = generate_forecast(
    sales, "region_a", train_pctg=0.75
)
forecast_a.plot()
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()

print(f"RMSE: {errors_a[0]:.2f} \nMAE: {errors_a[1]:.2f} \nMAPE: {errors_a[2]:%}")
```

```python
# Lets try to model another region as well
forecast_c, importance_plot_c, errors_c = generate_forecast(
    sales, "region_c", train_pctg=0.75
)
forecast_c.plot()
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()

print(f"RMSE: {errors_c[0]:.2f} \nMAE: {errors_c[1]:.2f} \nMAPE: {errors_c[2]:%}")
```
